* 'Ad Topic Line': Headline of the advertisement

* 'Country': Country of consumer

# Logistic Regression

Now it's time to do a train test split, and train our model!

You'll have the freedom here to choose columns that you want to train on!

** Split the data into training set and testing set using train_test_split**

```python
from sklearn.preprocessing import StandardScaler
scalar_ = StandardScaler()
```

```python
ad_data['Age'] = scalar_.fit_transform(ad_data[['Age']])
ad_data['Clicked on Ad'] = scalar_.fit_transform(ad_data[['Clicked on Ad']])
ad_data
```

```python
X=ad_data['Age']
Y = ad_data['Clicked on Ad']
ad_data['Age'].values.reshape(-1,1)
X = ad_data['Age'].values.reshape(-1,1)
y = ad_data['Clicked on Ad'].values.reshape(-1,1)
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = \
train_test_split(X, Y, test_size = 0.33, random_state = 42)
```

** Train and fit a logistic regression model on the training set.**

```python

from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
lr.fit(X_train, Y_train)
```

## Predictions and Evaluations
** Now predict values for the testing data.**

```python
Y_predict = lr.predict(X_test)
```

** Create a classification report for the model.**

```python
from sklearn import metrics
print("Classification Report for Decision Tree:""\n",metrics.classification_report(Y_test, Y_predict))
```

## Show the accuracy score

```python
print("Accuracy:",metrics.accuracy_score(Y_test, Y_predict))
```
