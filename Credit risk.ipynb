### 4.Impute missing values with mean (numerical variables)

```python
train.fillna(train.mean(),inplace=True)
```

### 5.Impute missing values with mode (categorical variables)

```python
train.Gender.fillna(train.Gender.mode(),inplace=True)
train.Married.fillna(train.Married.mode(),inplace=True)
train.Education.fillna(train.Education.mode(),inplace=True)
train.Self_Employed.fillna(train.Self_Employed.mode(),inplace=True)
```

```python
train
```

```python
test.Gender.fillna(test.Gender.mode(),inplace=True)
test.Married.fillna(test.Married.mode(),inplace=True)
test.Education.fillna(test.Education.mode(),inplace=True)
test.Self_Employed.fillna(test.Self_Employed.mode(),inplace=True)
```

```python
test
```

### PREDICTIVE MODELLING

### 6.Remove Loan_ID variable - Irrelevant

```python
train = train.drop("Loan_ID",axis=1)
test = test.drop("Loan_ID",axis=1)
```

### 7.Create target variable

```python
X = train.drop('Loan_Status',1)
Y = train.Loan_Status
```

### 8.Build dummy variables for categorical variables

```python
X = pd.get_dummies(X)
```

```python
train = pd.get_dummies(train)
test = pd.get_dummies(test)
```

### 9.Split train data for cross validation

```python
from sklearn.model_selection import train_test_split
X_train,X_cv,Y_train,Y_cv = train_test_split(X,Y,test_size=0.2)
```

### (a)LOGISTIC REGRESSION ALGORITHM

### 10.Fit model

```python
from sklearn.linear_model import LogisticRegression
model=LogisticRegression()
model.fit(X_train,Y_train)
```

### 11.Predict values for cv data

```python
pred_cv=model.predict(X_cv)
pred_cv
```

### 12.Print classification report

```python
from sklearn.metrics import classification_report
print(classification_report(Y_cv,pred_cv))
```

### 13.Evaluate accuracy of model

```python
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
```

```python
ac=accuracy_score(Y_cv,pred_cv) 
matrix=confusion_matrix(Y_cv,pred_cv)
print(ac)
print(matrix)
```

### (b)DECISION TREE ALGORITHM

### 14.Fit model

```python
from sklearn import tree
dt=tree.DecisionTreeClassifier(criterion= 'entropy')
dt.fit(X_train,Y_train)
```

### 15.Predict values for cv data

```python
pred_cv1=dt.predict(X_cv)
```

### 16.Evaluate accuracy of model

```python
ac1=accuracy_score(Y_cv,pred_cv1) 
matrix1=confusion_matrix(Y_cv,pred_cv1)
print(ac1)
print(matrix1)
```

### (c)SUPPORT VECTOR MACHINE (SVM) ALGORITHM

### 17.Fit model

```python
from sklearn import svm
svm_model=svm.SVC()
svm_model.fit(X_train,Y_train)
```

### 18.Predict values for cv data

```python
pred_cv2=svm_model.predict(X_cv)
```

### 19.Evaluate accuracy of model

```python
ac2=accuracy_score(Y_cv,pred_cv2) 
matrix2=confusion_matrix(Y_cv,pred_cv2)
print(ac2)
print(matrix2)
```

### (d)NAIVE BAYES ALGORITHM

### 20.Fit model

```python
from sklearn.naive_bayes import GaussianNB 
nb=GaussianNB()
nb.fit(X_train,Y_train)
```

### 21.Predict values for cv data

```python
pred_cv3=nb.predict(X_cv)
```

### 22.Evaluate accuracy of model

```python
ac3=accuracy_score(Y_cv,pred_cv3) 
matrix3=confusion_matrix(Y_cv,pred_cv3)
print(ac3)
print(matrix3)
```
