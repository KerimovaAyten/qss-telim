### 6.Construct AveragePrice distribution.

```python
plt.hist(df[df["region"]=="Albany"].AveragePrice,bins=30,fc=(1,0,0,0.5),label="Albany")
plt.hist(df[df["region"]=="WestTextNewMexico"].AveragePrice,bins=30,fc=(1,0,0.5),label="WestTextNewMexico")
plt.legend()
```

### 7.Which type of Avocados are expensive?(with boxplot)

```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
data = df
df= pd.read_csv('avocado.csv')
sns.boxplot(x="type", y="AveragePrice", data=df,palette='rainbow')
```

### 8.Construct heatmap correlation matrix.

```python
import seaborn as sns
f,ax=plt.subplots(figsize = (18,18))
sns.heatmap(df.corr(),annot= True,linewidths=0.5,fmt = ".1f")
plt.xticks(rotation=90)
plt.yticks(rotation=0)
plt.title('Correlation Map')
plt.savefig('graph.png')
plt.show()
```

### 9.Find relationship between large and small bags using scatterplot.(use jointplot function)

```python
sns.jointplot(x='Small Bags',y='Large Bags',data=df,kind='scatter')
```

### 10.Find linear relationship between Total Volume and Bags for each avocado type.(use lmplot function)

```python
sns.lmplot(x='Total Volume',y='Small Bags',data=df,hue='type')
sns.lmplot(x='Total Volume',y='Large Bags',data=df,hue='type')
sns.lmplot(x='Total Volume',y='XLarge Bags',data=df,hue='type')

```
